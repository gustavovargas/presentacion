- Loans [:, 0] 'id': A unique LC assigned ID for the loan listing.

- Loans [:, 1] 'member_id': A unique LC assigned Id for the borrower member.
- Loans [:, 2] 'loan_amnt': The listed amount of the loan applied for by the borrower. If at some point in time, the credit department reduces the loan amount, then it will be reflected in this value.
- Loans [:, 3] 'funded_amnt': The total amount committed to that loan at that point in time.
- Loans [:, 4] 'funded_amnt_inv': The total amount committed by investors for that loan at that point in time.
- Loans [:, 5] 'term': The number of payments on the loan. Values are in months and can be either 36 or 60.
- Loans [:, 6] 'int_rate': Interest Rate on the loan
- Loans [:, 7] 'installment': The monthly payment owed by the borrower if the loan originates.
- Loans [:, 8] 'grade': LC assigned loan grade
- Loans [:, 9] 'sub_grade': LC assigned loan subgrade
- Loans [:, 10] 'emp_title': The job title supplied by the Borrower when applying for the loan.
- Loans [:, 11] 'emp_length': Employment length in years. Possible values are between 0 and 10 where 0 means less than one year and 10 means ten or more years.
- Loans [:, 12] 'home_ownership': The home ownership status provided by the borrower during registration. Our values are: RENT, OWN, MORTGAGE, OTHER.
- Loans [:, 13] 'annual_inc': The self-reported annual income provided by the borrower during registration.
- Loans [:, 14] 'verification_status': Indicates if the co-borrowers' joint income was verified by LC, not verified, or if the income source was verified
- Loans [:, 15] 'issue_d': The month which the loan was funded
- Loans [:, 16] 'loan_status': Current status of the loan
- Loans [:, 17] 'pymnt_plan': Indicates if a payment plan has been put in place for the loan
- Loans [:, 18] 'url': URL for the LC page with listing data.
- Loans [:, 19] 'desc': Loan description provided by the borrower
- Loans [:, 20] 'purpose': A category provided by the borrower for the loan request. 
- Loans [:, 21] 'title': The loan title provided by the borrower
- Loans [:, 22] 'zip_code': The first 3 numbers of the zip code provided by the borrower in the loan application.
- Loans [:, 23] 'addr_state': The state provided by the borrower in the loan application
- Loans [:, 24] 'dti': A ratio calculated using the borrower’s total monthly debt payments on the total debt obligations, excluding mortgage and the requested LC loan, divided by the borrower’s self-reported monthly income.
- Loans [:, 25] 'delinq_2yrs': The number of 30+ days past-due incidences of delinquency in the borrower's credit file for the past 2 years
- Loans [:, 26] 'earliest_cr_line': The month the borrower's earliest reported credit line was opened
- Loans [:, 27] 'inq_last_6mths': The number of inquiries in past 6 months (excluding auto and mortgage inquiries)
- Loans [:, 28] 'mths_since_last_delinq': The number of months since the borrower's last delinquency.
- Loans [:, 29] 'mths_since_last_record': The number of months since the last public record.
- Loans [:, 30] 'open_acc': The number of open credit lines in the borrower's credit file.
- Loans [:, 31] 'pub_rec': Number of derogatory public records
- Loans [:, 32] 'revol_bal': Total credit revolving balance
- Loans [:, 33] 'revol_util': Revolving line utilization rate, or the amount of credit the borrower is using relative to all available revolving credit.
- Loans [:, 34] 'total_acc': The total number of credit lines currently in the borrower's credit file
- Loans [:, 35] 'initial_list_status': The initial listing status of the loan. Possible values are – W, F
- Loans [:, 36] 'out_prncp': Remaining outstanding principal for total amount funded
- Loans [:, 37] 'out_prncp_inv': Remaining outstanding principal for portion of total amount funded by investors
- Loans [:, 38] 'total_pymnt': Payments received to date for total amount funded
- Loans [:, 39] 'total_pymnt_inv': Payments received to date for portion of total amount funded by investors
- Loans [:, 40] 'total_rec_prncp': Principal received to date
- Loans [:, 41] 'total_rec_int': Interest received to date
- Loans [:, 41] 'total_rec_int': Interest received to date
- Loans [:, 42] 'total_rec_late_fee': Late fees received to date
- Loans [:, 43] 'recoveries': post charge off gross recovery
- Loans [:, 44] 'collection_recovery_fee': post charge off collection fee
- Loans [:, 45] 'last_pymnt_d': Last month payment was received
- Loans [:, 46] 'last_pymnt_amnt': Last total payment amount received
- Loans [:, 47] 'next_pymnt_d': Next scheduled payment date
- Loans [:, 48] 'last_credit_pull_d': The most recent month LC pulled credit for this loan
- Loans [:, 49] 'collections_12_mths_ex_med': Number of collections in 12 months excluding medical collections
- Loans [:, 50] 'mths_since_last_major_derog': Months since most recent 90-day or worse rating
- Loans [:, 51] 'policy_code: publicly available policy_code=1
new products not publicly available policy_code=2
- Loans [:, 52] 'application_type': Indicates whether the loan is an individual application or a joint application with two co-borrowers
- Loans [:, 53] 'annual_inc_joint': The combined self-reported annual income provided by the co-borrowers during registration
- Loans [:, 54] 'dti_joint': A ratio calculated using the co-borrowers' total monthly payments on the total debt obligations, excluding mortgages and the requested LC loan, divided by the co-borrowers' combined self-reported monthly income
- Loans [:, 55] 'verification_status_joint': Indicates if the co-borrowers' joint income was verified by LC, not verified, or if the income source was verified
- Loans [:, 56] 'acc_now_delinq': The number of accounts on which the borrower is now delinquent.
- Loans [:, 57] 'tot_coll_amt': Total collection amounts ever owed
- Loans [:, 57] 'tot_cur_bal': Total current balance of all accounts
- Loans [:, 58] 'open_acc_6m': Number of open trades in last 6 months
- Loans [:, 59] 'open_il_6m': Number of currently active installment trades
- Loans [:, 60] 'open_il_12m': Number of installment accounts opened in past 12 months
- Loans [:, 61] 'open_il_24m': Number of installment accounts opened in past 24 months
- Loans [:, 62] 'mths_since_rcnt_il': Months since most recent installment accounts opened
- Loans [:, 63] 'total_bal_il': Total current balance of all installment accounts
- Loans [:, 64] 'il_util': Ratio of total current balance to high credit/credit limit on all install acct
- Loans [:, 65] 'open_rv_12m': Number of revolving trades opened in past 12 months
- Loans [:, 66] 'open_rv_24m': Number of installment accounts opened in past 24 months
- Loans [:, 67] 'max_bal_bc': Maximum current balance owed on all revolving accounts
- Loans [:, 68] 'all_util': Balance to credit limit on all trades
- Loans [:, 69] 'total_rev_hi_lim': Total revolving high credit/credit limit
- Loans [:, 70] 'inq_fi': Number of personal finance inquiries
- Loans [:, 71] 'total_cu_tl': Number of finance trades
- Loans [:, 72] 'inq_last_12m': Number of credit inquiries in past 12 months
